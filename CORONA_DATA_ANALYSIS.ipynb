```python
import matplotlib.pyplot as plt
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])

# Group by 'ObservationDate' and sum the 'Confirmed' cases globally
global_confirmed = df.groupby('ObservationDate')['Confirmed'].sum()

# Plot the trend of confirmed cases over time
plt.figure(figsize=(10, 6))
plt.plot(global_confirmed.index, global_confirmed.values, color='b', label='Confirmed Cases')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.title('Trend of Confirmed Cases Globally Over Time')
plt.legend()

# Show plot
plt.show()
```

# Which province/state has reported the highest number of confirmed cases?

```python
# Group by 'Province/State' and sum the 'Confirmed' cases
state_confirmed = df.groupby('Province/State')['Confirmed'].sum()

max_confirmed_state = state_confirmed.idxmax()
max_confirmed_cases = state_confirmed.max()

print(f"The province/state with the highest confirmed cases is {max_confirmed_state} with {max_confirmed_cases} cases.")
```

# How does the mortality rate vary across different countries/regions? 

```python
# Group data by 'Country/Region' and sum confirmed cases and deaths
country_stats = df.groupby('Country/Region').agg({'Confirmed': 'sum', 'Deaths': 'sum'}).reset_index()

# Calculate mortality rate
country_stats['Mortality Rate (%)'] = (country_stats['Deaths'] / country_stats['Confirmed']) * 100

# Display the updated DataFrame
country_stats
```

```python
import matplotlib.pyplot as plt

# Assuming country_stats is your DataFrame containing the mortality rates

# Sort by Mortality Rate for better visualization
country_stats = country_stats.sort_values(by='Mortality Rate (%)', ascending=False)
country_stats = country_stats[country_stats['Mortality Rate (%)'] > 0]
# Create horizontal bar chart
plt.figure(figsize=(12, 25))
plt.barh(country_stats['Country/Region'], country_stats['Mortality Rate (%)'], color='skyblue')

# Add labels and title
plt.title('Mortality Rate (%) by Country/Region')
plt.xlabel('Mortality Rate (%)')
plt.ylabel('Country/Region')

# Show the plot
plt.grid(axis='x', alpha=0.75)  # Optional: add grid for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

# How does the recovery rate vary across different countries/regions?

```python
country_stats = df.groupby('Country/Region').agg({
    'Confirmed': 'sum',
    'Recovered': 'sum',
    'Deaths': 'sum'}).reset_index()

# Calculate recovery rate
country_stats['Recovery Rate (%)'] = (country_stats['Recovered'] / country_stats['Confirmed']) * 100


# Sort by Recovery Rate for better visualization
country_stats = country_stats.sort_values(by='Recovery Rate (%)', ascending=False)
country_stats = country_stats[country_stats['Recovery Rate (%)'] > 0]
# Create horizontal bar chart for recovery rates
plt.figure(figsize=(12, 25))
plt.barh(country_stats['Country/Region'], country_stats['Recovery Rate (%)'], color='lightgreen')

plt.title('Recovery Rate (%) by Country/Region')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Country/Region')

# Show the plot
plt.grid(axis='x', alpha=0.75)  # Optional: add grid for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python

# Sort the DataFrame by date
df.sort_values('ObservationDate', inplace=True)

# Group by 'ObservationDate' and sum confirmed cases to get daily new cases
df['New Confirmed'] = df.groupby('Country/Region')['Confirmed'].diff().fillna(0)

# Sum new confirmed cases for each date globally
global_new_cases = df.groupby('ObservationDate')['New Confirmed'].sum().reset_index()

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(global_new_cases['ObservationDate'], global_new_cases['New Confirmed'], color='blue', linewidth=2)

# Add labels and title
plt.title('Trend of New Confirmed Cases Over Time Globally')
plt.xlabel('Date')
plt.ylabel('New Confirmed Cases')
plt.xticks(rotation=45)
plt.grid()

# Show the plot
plt.show()
```

```python

province_stats = df.groupby('Province/State').agg({'Confirmed': 'sum', 'Deaths': 'sum'}).reset_index()

# Calculate fatality rate
province_stats['Fatality Rate (%)'] = (province_stats['Deaths'] / province_stats['Confirmed']) * 100


# Filter out provinces/states with a fatality rate of 0%
province_stats = province_stats[province_stats['Fatality Rate (%)'] > 0]

province_stats.sort_values('Fatality Rate (%)', ascending=False, inplace=True)


plt.figure(figsize=(12, 30))
plt.barh(province_stats['Province/State'], province_stats['Fatality Rate (%)'], color='blue')

# Add labels and title
plt.title('Fatality Rate by Province/State (Excluding 0% Rates)')
plt.xlabel('Fatality Rate (%)')
plt.ylabel('Province/State')
plt.grid(axis='x')
plt.tight_layout()
# Show the plot
plt.show()
```

# the trend of active cases over time globally

```python
df['Active Cases'] = df['Confirmed'] - df['Deaths'] - df['Recovered']

# Group by date and sum active cases
daily_active_cases = df.groupby('ObservationDate')['Active Cases'].sum().reset_index()

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(daily_active_cases['ObservationDate'], daily_active_cases['Active Cases'], color='blue', marker='o', linewidth=2)

# Add labels and title
plt.title('Global Trend of Active Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Active Cases')
plt.xticks(rotation=45)
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```
